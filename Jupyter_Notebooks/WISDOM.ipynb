```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data1=pd.read_csv('Documents/Python.csv')
```

```python
jet=data1[data1['Radio Jet']=='Yes']
nojet=data1[data1['Radio Jet']=='No']
```

```python
plt.hist(jet['Spectral Index'],color='Red',label='Jet',alpha=0.5)
plt.hist(nojet['Spectral Index'],color='Blue',label='No Jet',alpha=0.5)
plt.xlabel('mm Spectral Index')
plt.ylabel('Count')
plt.legend()
```
