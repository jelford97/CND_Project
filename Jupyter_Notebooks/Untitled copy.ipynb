```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data=pd.read_csv('/Users/jelford/Documents/PhD_Work/CND_Project/CSV_Files/Python.csv')
mm=data['mm Luminosity (erg/s)']
radio=data['Radio Luminosity (erg/s)']
xray=data['X-ray Luminosity (ergs/s)'] 
```

```python
fig=plt.figure()
ax1=plt.subplot(223)
plt.loglog(mm,xray,'o')
plt.xlabel('Nuclear mm Luminosity (erg/s)')
plt.ylabel('Global X-ray \nLuminosity \n(erg/s)')
plt.subplots_adjust(hspace=0)
ax2=plt.subplot(221,sharex=ax1)
plt.tick_params('x', labelbottom=False)
plt.ylabel('Global Radio \nLuminosity \n(erg/s)')
plt.loglog(mm,radio,'o')
ax3=plt.subplot(224,sharey=ax1)
plt.tick_params('y', labelleft=False)
plt.subplots_adjust(wspace=0)
plt.loglog(radio,xray,'o')
plt.xlabel('Global Radio Luminosity (erg/s)')
```
