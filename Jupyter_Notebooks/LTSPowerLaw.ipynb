```python
from ltsfit.lts_linefit import lts_linefit
import pandas as pd
import numpy as np
import matplotlib as plt
```

```python
data=pd.read_csv('Documents/Python.csv')
```

```python
df=data[['Mass 50pc','Mass 50pc Error','Radio Flux (Jy)','Radio Flux Uncertainty (Jy)','Radio Luminosity (L_solar)']]
dff=df.dropna()
mass=dff['Mass 50pc']
masserr=dff['Mass 50pc Error']
radio=dff['Radio Flux (Jy)']
radioerr=dff['Radio Flux Uncertainty (Jy)']
radiolum=dff['Radio Luminosity (L_solar)']*3.846e33
errscale=radioerr/radio
radioLerr=radiolum*errscale
radioL=np.array(radiolum)
massA=np.array(mass)
masserrA=np.array(masserr)
radioerrA=np.array(radioLerr)
```

```python
np.polyfit(np.log10(massA),np.log10(radioL),1)
```

```python
lts_linefit(np.log10(massA),np.log10(radioL),(masserrA/massA)*0.434,(radioerrA/radioL)*0.434, pivot=np.median(np.log10(massA)))
```
