```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def luminosity(lbol):
    x=np.log10(lbol)
    y=x-0.24*(x-12)-0.012*((x-12)**2)+0.0015*((x-12)**3)-1.54
    lxray=10**(y)
    return lxray
```

```python
lbol=np.logspace(5.5,6,1000)
```

```python
xrayL=[]
for i in range(len(lbol)):
    xray=luminosity(lbol[i])
    xrayL.append(xray)
```

```python
xrayLnew=np.array(xrayL)
lbolnew=np.array(lbol)
```
