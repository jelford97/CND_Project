```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
plt.rcParams['font.family'] = 'Latin Modern Roman'
plt.rcParams.update({'font.size': 10})
plt.rcParams['legend.fontsize'] = 12
plt.rcParams['axes.linewidth'] = 1.5
plt.rcParams['xtick.labelsize'] = 12;
plt.rcParams['ytick.labelsize'] = 12
plt.rcParams['xtick.major.size'] = 10;
plt.rcParams['ytick.major.size'] = 10
plt.rcParams['xtick.major.width'] = 2;
plt.rcParams['ytick.major.width'] = 2
plt.rcParams['xtick.minor.size'] = 5;
plt.rcParams['ytick.minor.size'] = 5
plt.rcParams['xtick.minor.width'] = 1;
plt.rcParams['ytick.minor.width'] = 1
plt.rcParams['xtick.direction'] = 'in';
plt.rcParams['ytick.direction'] = 'in'
plt.rcParams['xtick.bottom'] = True
plt.rcParams['ytick.left'] = True
params = {'mathtext.default': 'regular'}
plt.rcParams.update(params)
plt.rcParams['axes.labelsize'] = 14
plt.rcParams['ytick.right']=True
plt.rcParams['xtick.top']=True
```

```python
data=pd.read_csv('Documents/Python.csv')
```

```python
liner=data[data['AGN Class']=='LINER']
seyfert=data[data['AGN Class']=='Seyfert']
seyfert2=data[data['AGN Class']=='Seyfert 2']
other=data[data['AGN Class']=='Other']
noclass=data[data['AGN Class']=='No class']
```

```python
plt.loglog(liner['Radio Flux (Jy)'],liner['X-ray flux density(Jy)'],'o',color='blue',label='LINER')
plt.loglog(seyfert['Radio Flux (Jy)'],seyfert['X-ray flux density(Jy)'],'o',color='red',label='Seyfert')
plt.loglog(seyfert2['Radio Flux (Jy)'],seyfert2['X-ray flux density(Jy)'],'o',color='green',label='Seyfert 2')
plt.loglog(other['Radio Flux (Jy)'],other['X-ray flux density(Jy)'],'o',color='yellow',label='Other')
plt.loglog(noclass['Radio Flux (Jy)'],noclass['X-ray flux density(Jy)'],'o',color='black',label='No Class')
plt.xlabel('radio flux density(Jy)')
plt.ylabel('X-ray flux density(Jy)')
plt.legend(bbox_to_anchor=(1, 0), loc='lower left', ncol=1)
```

```python
jet=data[data['Radio Jet']=='Yes']
nojet=data[data['Radio Jet']=='No']
```

```python
plt.loglog(jet['Radio Flux (Jy)'],jet['X-ray flux density(Jy)'],'o',color='blue',label='Jet')
plt.loglog(nojet['Radio Flux (Jy)'],nojet['X-ray flux density(Jy)'],'o',color='red',label='No Jet')
plt.xlabel('radio flux density(Jy)')
plt.ylabel('x ray density(Jy)')
plt.legend(bbox_to_anchor=(1, 0), loc='lower left', ncol=1)
```

```python
linerdect=liner[liner['Detection']=='Yes']
seyfert2dect=seyfert2[seyfert2['Detection']=='Yes']
otherdect=other[other['Detection']=='Yes']
noclassdect=noclass[noclass['Detection']=='Yes']
linernodect=liner[liner['Detection']=='No']
seyfert2nodect=seyfert2[seyfert2['Detection']=='No']
othernodect=other[other['Detection']=='No']
noclassnodect=noclass[noclass['Detection']=='No']
```

```python
plt.loglog(linerdect['1mm flux density (Jy)'],linerdect['X-ray flux density(Jy)'],'o', color='blue',label='LINER')
plt.loglog(seyfert['1mm flux density (Jy)'],seyfert['X-ray flux density(Jy)'],'o',color='red',label='Seyfert')
plt.loglog(seyfert2dect['1mm flux density (Jy)'],seyfert2dect['X-ray flux density(Jy)'],'o',color='green',label='Seyfert 2')
plt.loglog(otherdect['1mm flux density (Jy)'],otherdect['X-ray flux density(Jy)'],'o',color='yellow',label='Other')
plt.loglog(noclassdect['1mm flux density (Jy)'],noclassdect['X-ray flux density(Jy)'],'o',color='black',label='No class')
plt.loglog(linernodect['1mm flux density (Jy)'],linernodect['X-ray flux density(Jy)'],'v',color='blue',label='LINER non-detection')
plt.loglog(noclassnodect['1mm flux density (Jy)'],noclassnodect['X-ray flux density(Jy)'],'v',color='black',label='No class non-detection')
plt.legend(bbox_to_anchor=(1, 0), loc='lower left', ncol=1)
plt.xlabel('mm flux density (Jy)')
plt.ylabel('X-ray flux density (Jy)')
```
