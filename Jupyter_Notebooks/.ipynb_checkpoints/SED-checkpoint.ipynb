```python
def mod_black_body(M,T,v,v0,D,b):
    d=D*3.0856e22
    m=M*2e30
    c=3e8
    h=6.626e-34
    kb=1.38e-23
    B=((2*h*v**3/c**2)*1/(np.exp(h*v/(kb*T))))
    k0=0.051
    k=k0*((v/v0)**b)
    S=(k*B*m/(d**2))
    sj=S/1e-26
    return sj
```

```python
def freefree(SFR,v,Te,dist):
    d=dist*3.086e22
    erg=SFR/(4.6e-28*(Te/1e4)**(-0.45)*(v/1e9)**0.1)
    Lum=erg/1e7
    sj=Lum/(4*math.pi*d**2)
    s=sj/1e-26
    return s
```

```python
def power_law(a,p,K):
    B=(K*(a/1e10)**p)
    return B
```

```python
def combined(p,K,M,T,v,v0,D,b):
    com=power_law(v,p,K)+mod_black_body(M,T,v,v0,D,b)
    return com
```

```python
def freefreecom(p,K,M,T,v,v0,D,b,SFR,Te):
    comff=power_law(v,p,K)+mod_black_body(M,T,v,v0,D,b)+freefree(SFR,v,Te,D)
    return comff
```

```python
def plotter(p,k,M,T,D,b,nucp,nuck,nucm,LSB,LSBfr,USB,USBfr,SEDfl,SEDfr):
    rfreq=np.logspace(7,12,100)
    dfreq=np.logspace(10,13.5,100)
    ffreq=np.logspace(7,13.5,100)
    plt.loglog(SEDfr,SEDfl,'o',color='blue')
    plt.loglog(LSBfr,LSB,'*',markersize=12,color='orange')
    plt.loglog(USBfr,USB,'*',markersize=12,color='green')
    pl=power_law(rfreq,p,k)
    dust=mod_black_body(M,T,dfreq,6e11,D,b)
    com=combined(p,k,M,T,ffreq,6e11,D,b)
    nuc=combined(nucp,nuck,nucm,T,ffreq,6e11,D,b)
    plt.loglog(dfreq,dust,color='green')
    plt.loglog(rfreq,pl,color='red')
    plt.loglog(ffreq,com,color='orchid')
    plt.loglog(ffreq,nuc,color='blue')
    plt.xlabel('Frequency (Hz)')
    plt.ylabel('Flux Density (Jy)')  
```

```python
def SEDplotter(mm,mmfreq,datamm,datafreq,dataerr,K,p,M,T,v0,D,b):
    rfreq=np.logspace(7,12,100)
    dfreq=np.logspace(10,13,100)
    ffreq=np.logspace(7,13,100)
    Rfit=power_law(rfreq,p,K)
    Dfit=mod_black_body(M,T,dfreq,v0,D,b)
    Cfit=combined(p,K,M,T,ffreq,v0,D,b)
    plt.loglog(mmfreq,mm,'*',markersize=12,color='orange')
    plt.loglog(datafreq,datamm,'o')
    plt.loglog(rfreq,Rfit,color='red')
    plt.loglog(dfreq,Dfit,color='green')
    plt.loglog(ffreq,Cfit,color='orchid')
    plt.errorbar(datafreq,datamm,yerr=dataerr,fmt='.',ecolor='black')
    plt.xlabel('Frequency (Hz)')
    plt.ylabel('Flux Density (Jy)')
    plt.xlim(1e6,1e13)
    plt.ylim(1e-5)
```
