```python
def radio_flux(SFR,Te,fr,dist):
    radiolum=(SFR/1e-27)*(2.18*(Te/1e4)**0.45*(fr/1e9)**(-0.1)+15.1*(fr/1e9)**-0.8)
    d=dist*3.086e24
    radio=radiolum/(4*math.pi*(d)**2)
    radioflux=radio/1e-23
    return radioflux
```

```python
def mmflux(v,d,T,v0,b,M):
    D=d*3.086e22
    m=M*2e30
    c=3e8
    h=6.626e-34
    kb=1.38e-23
    k0=0.051
    k=k0*(v/v0)**b
    B=(2*h*v**3/c**2)*(1/(np.exp(h*v/(kb*T))-1))
    S=k*m*B/D**2
    sj=S/1e-26
    return sj
```
